Now lets eradicate the outliers present in the datasets ie in itemoutlet sell and outlet visibiltiy.

```python
df.describe()
```

```python
# Firstly finding IOR for itemOutlet sell
percentile25=df['Item_Outlet_Sales'].quantile(0.25)
percentile75=df['Item_Outlet_Sales'].quantile(0.75)
```

```python
percentile75
```

```python
percentile25
```

```python
IQR=percentile75-percentile25
```

```python
IQR
```

```python
#finding upperlimit and lower limit
upperlimit=percentile75+1.5*IQR
lowerlimit=percentile25-1.5*IQR
```

```python
upperlimit
```

```python
lowerlimit
```

Therefore we conclude that any values above upperlimit(3251.0887839375914) and lowerlimit(398.2891296374)
are outlier                                                                                       

We can eliminate it by using using two method ie capping and clipping but we are using capping method.

```python
new_df=df.copy()
```

```python
new_df['Item_Outlet_Sales']=np.where(new_df['Item_Outlet_Sales']>upperlimit,upperlimit,
        np.where(new_df['Item_Outlet_Sales']<lowerlimit,lowerlimit,new_df['Item_Outlet_Sales']))
```

```python
new_df.describe()
```

```python
sns.boxplot(x='Item_Outlet_Sales',data=new_df)
plt.show()
```
