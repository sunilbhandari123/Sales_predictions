```python
df.isnull().sum()
```

Here we will drop item identifier and outlet identifier because i thik it is not related to the price of the product

```python
df=df.drop(columns=['Item_Identifier','Outlet_Identifier'],axis=0)
```

```python
df.head()
```

Here we can see Itemweight  outlet size and item outlet size have the missing values among them 2 are 
of float type and onr outlet size is object type.  we will replace the missing values by its median values int float data type

```python
df['Item_Weight']=df.Item_Weight.fillna(df.Item_Weight.median())
```

```python
df['Item_Outlet_Sales']=df.Item_Outlet_Sales.fillna(df.Item_Outlet_Sales.mean())
```
